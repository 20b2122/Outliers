```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
filename = './data/BostonAirbnb/listings.csv'
df=pd.read_csv(filename)
display(df[['property_type','price']].head(5))
```

# Hampel Method
* Find the median of the dataset.
* Calculate the absolute deviation of each data point from the median.
* Calculate the median of the deviations.
* Check the absolute deviation against the value of 4.5*median of the deviations.
* Whichever data point is greater or equal to that critical value, is considered as outlier.

```python
def remove_outlier_Hampel(df):
    median = df.median()
    List = abs(df-median)
    cond = List.median()*4.5
    good_list = List[~(List>cond)]
    return good_list
```

#### >removing the '$' sign to convert the price as float

```python
def remove_sign(x,sign):
    if type(x) is str:
        x = float(x.replace(sign,'').replace(',',''))
    return x
```

```python
df=df[['price','property_type']]
df=pd.DataFrame(df)

df.price = df.price.apply(remove_sign,sign='$')
sns.boxplot(y='price', x='property_type',data=df);
plt.xticks(rotation=90);
plt.ylabel('Price ($)');
```

```python
df_outlier_removed = remove_outlier_Hampel(df.price)
df_outlier_removed = pd.DataFrame(df_outlier_removed)
ind_diff = df.index.difference(df_outlier_removed.index)

for i in range (0, len(ind_diff), 1):
    df_final = df.drop([ind_diff[i]])
    df = df_final
    
sns.boxplot(y='price', x='property_type', data=df_final);
plt.xticks(rotation=90);
plt.ylabel('Price ($)');
```
